# Area of Pi Code

```
import numpy as np
import matplotlib.pyplot as plt
import random

inside_circle = 0

# n is the number of random sample points that will be used...
n = 10000 #10000 #100 #1000 #5000 

x_inside = []
y_inside = []
x_outside = []
y_outside = []

# create for loop to control the random sampling generator
for i in range(n):
  # be able to sample the x and y values from the uniform distribution
    x = np.random.uniform(-1.2,1.2)
    y = np.random.uniform(-1.2,1.2)

    # create acceptance and rejection of points based on the area 
    # of the circle (whether inside or not)
    if x**2 + y**2 <= 1:
        inside_circle += 1
        x_inside.append(x)
        y_inside.append(y)
    else:
        x_outside.append(x)
        y_outside.append(y)


# create the approximation of pi using the accepted points in the circle...
# multiplied by 4 to create full circle approx. not just a single quadrant...

est_pi = 4*inside_circle/n
accepted_pi = np.pi 
print('Accepted Pi = ', accepted_pi, 'Estimated Pi = ', est_pi)

percent_error = 100 * abs(accepted_pi - est_pi)/ accepted_pi
print('The Percent Error is = ', percent_error, '%')


plt.figure (figsize = [5,5])
plt.title ('The Area of Pi Monte Carlo Simulation')
plt.scatter(x_inside, y_inside, color='blue', s=10)
plt.scatter(x_outside, y_outside, color='pink', s=10)
plt.show()

```
